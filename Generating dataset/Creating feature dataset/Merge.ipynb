```python
import pandas as pd
import numpy as np
```

```python
reliable_features = pd.read_csv('../reliable_features.csv')
reliable_sparse_wwe = pd.read_csv('../reliable_sparse_withwideeyes_features.csv')
reliable_800_wwe = pd.read_csv('../reliable_withwideeyes_features.csv')
A = [reliable_features,reliable_sparse_wwe,reliable_800_wwe]

color_features = pd.read_csv('color_features.csv')
composition_features = pd.read_csv('composition_features.csv')
B = [color_features,composition_features]
final = []
name = ['human_features.csv', 'human_features_sparse_wwe.csv', 'human_features_800_wwe.csv']
for i in A:
    s = i
    for j in B:
        s = pd.merge(s,j, on = 'product_id')
    final.append(s)

for i,content in enumerate(final):
    display(content.head())
    display(content.shape)
    content.to_csv(name[i], index = False)
```
