```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
composition = pd.read_csv('composition.csv')
display(composition)
```

```python
composition[composition['composition']==np.nan]
```

```python
## Pie chart
plt.figure(figsize=(12,12))
# plt.title("Distribution of season",fontsize = 11)
sorted_counts = composition['composition'].value_counts()
print(sorted_counts[:20])
# plt.pie(x=sorted_counts,labels=sorted_counts.index,autopct="%0.2f%%")
# plt.savefig("../../saved_figs/Pie chart of composition.pdf",dpi=800)
plt.show()
```

```python
composition['composition'] = composition.composition.str.split(',')
composition = composition.set_index(['product_id'])['composition'].apply(pd.Series).stack().reset_index(level=1, drop=True)
composition = composition.reset_index()
composition.columns = ['product_id', 'composition']
composition[['percentage', 'material']] = pd.DataFrame(composition.composition.str.split('% ', 1).tolist(), columns = ['percentage', 'material'])
composition = composition.pivot(index='product_id', columns='material', values='percentage').reset_index()
composition.columns.name = None
composition.fillna(0, inplace=True)
display(composition)
```
